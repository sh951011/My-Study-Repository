```python
%%time
import torch
import torch.nn as nn
from torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader
import torch.nn.init as init
import torchvision.transforms as transforms
import librosa
import librosa.display
import numpy as np

MFCC_X = 20
MFCC_Y = 44
batch_size     =   10
learning_rate  =   0.0003
epoch          =   10
```

```python
def _path(loc, directory, filename, redundant = "", format_ = ".wav"):
    delimiter = '\\'
    path = ( loc + delimiter + directory + delimiter +
           filename + redundant + format_)
    return path
```

```python
def _normalize(mfcc):
    return (mfcc - np.min(mfcc)) / (np.max(mfcc) - np.min(mfcc))
```

```python
def _fit_size(arr, size):
    zeros = np.zeros( (len(arr), size - len(arr[0]) ) )
    arr = np.append(arr, zeros, axis = 1)
    return arr
```

```python
def _mfcc(path):
    y, sr = librosa.load(path)
    mfcc = librosa.feature.mfcc(y = y, sr = sr)
    mfcc = _fit_size(mfcc, MFCC_Y)
    return  _normalize(mfcc)
```

```python
class WordDataset(Dataset):
    def __init__(self, train = True):
        label_list = [x for x in range(30)]
        self.label_word = ['bed', 'bird', 'cat', 'dog', 'down', 'eight', 'five',
                          'four', 'go', 'happy', 'house', 'left', 'marvin', 'nine',
                          'no', 'off', 'on', 'one', 'right', 'seven', 'sheila',
                          'six', 'stop', 'three', 'tree', 'two', 'up', 'wow', 'yes', 'zero']
        self.train = train
        num_data   =  1700
        self.num_train  =  1600
        self.num_test   =  100

        if self.train == True: 
            self.train_data   =  []
            self.train_label  =  []
            
            print("\n\n==== Train Data:")
            for item in label_list:
                print("===", end = "")
                for i in range(1, self.num_train + 1):
                    path = _path(loc = '..\\테스트\\dataset', 
                              directory = self.label_word[item],filename = self.label_word[item],
                                 redundant = ' (' + str(i) + ')', format_ = '.wav')
                    mfcc = _mfcc(path)
                    self.train_data.append(mfcc)
                    self.train_label.append(item)
                    
            self.train_label = np.array(self.train_label)
            self.train_data = np.concatenate(self.train_data)
            self.train_data = self.train_data.reshape(self.num_train * len(label_list), 1, MFCC_X, MFCC_Y)
            print("=== Dataset Download Complete !!")
            
        else:
            self.test_data   =  []
            self.test_label  =  []
            
            print("\n\n=== Test Data:")
            for item in label_list:
                print("===", end = "")
                for i in range(self.num_train + 1, num_data + 1):              
                    path = _path(loc = '..\\테스트\\dataset', 
                              directory = self.label_word[item],filename = self.label_word[item],
                                 redundant = ' (' + str(i) + ')', format_ = '.wav')
                    mfcc = _mfcc(path)
                    self.test_data.append(mfcc)
                    self.test_label.append(item)
                    
            self.test_label = np.array(self.test_label)
            self.test_data = np.concatenate(self.test_data)
            self.test_data = self.test_data.reshape(num_test * len(label_list), 1, MFCC_X, MFCC_Y)
            print("=== Dataset Download Complete !!")
        
    def __getitem__(self, index):
        if self.train:
            return self.train_data[index], self.train_label[index]
        else:
            return self.test_data[index], self.test_label[index]
    
    def __len__(self):
        if self.train:
            return len(self.train_data)
        else:
            return len(self.test_data)
```

```python
class CNN(nn.Module):
    def __init__(self):
        #  CNN 클래스의 부모 클래스인 nn.Module을 초기화하는 역할
        super(CNN, self).__init__() 
        
        self.layer = nn.Sequential(
            nn.Conv2d(1,16,5),
            nn.ReLU(),
            nn.Conv2d(16,32,5),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            
            # [batch_size, 32, 20, 20] => [batch_size, 32, 10, 10]
            
            nn.Conv2d(32,64,5),
            nn.ReLU(),
            nn.MaxPool2d(2,2)
            
            # [batch_size, 32, 10, 10] => [batch_size, 32, 6, 6] => [batch_size, 64, 3, 3]
        )
        
        # Linear 함수를 통해 뉴런의 수를 줄여나간다.
        self.fc_layer = nn.Sequential(
            nn.Linear(64*1*7, 100),
            nn.ReLU(),
            nn.Linear(100, 30)
        )
        # Output : [batch_size, 10] 형태의 텐서
        
    def forward(self, x):
        out = self.layer(x)
        out = out.view(batch_size, -1)  # 여기서 -1은 -1인 부분은 알아서 계산하라는 의미 (reshape()와 같음)
        out = self.fc_layer(out)
        return out 
```

```python
%%time
test_dataset = WordDataset(train = False)
train_dataset = WordDataset(train = True)

train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle = True, num_workers = 0, drop_last = True)
test_loader = DataLoader(test_dataset, batch_size = batch_size, shuffle = True, num_workers = 0, drop_last = True)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = CNN().to(device)
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate) # SGD가 아닌 Adam 알고리즘 사용
```

```python
%%time
loss_arr = []

for i in range(epoch):
    cursor = 0
    for j, [image,label] in enumerate(train_loader):
        x = image.to(device)
        y_ = label.to(device)
        
        y_ = torch.tensor(y_).long()
        
        if j == 0:
            if i == 0:
                print("Train on 48000 samples, validate on 3000 samples:")
            print("Epoch " , (i+1) , "/ 4:")
            print("[", end = "")
            
        model = model.double()
        
        optimizer.zero_grad()
        output = model.forward(x)
        loss = loss_func(output, y_)
        loss.backward()
        optimizer.step()
        
        if (j + 1) % 480 == 0 :
            cursor += 1
            print("==", end = "")
    print("=]")    
    print(loss)                                                
    loss_arr.append(loss.cpu().detach().numpy())

import matplotlib.pyplot as plt
plt.plot(loss_arr)
plt.xlabel('$epoch$')
plt.ylabel('$loss$')
plt.grid(True)
plt.show
```

```python
correct = 0
total = 0

with torch.no_grad():
    for audio, label in test_loader:
        x = audio.to(device)
        y_ = label.to(device)
        
        y_ =  torch.tensor(y_).long()
        output = model.forward(x)
        _,output_index = torch.max(output, 1)
        
        total += label.size(0)                                          
        correct += (output_index == y_).sum().float()
        
    print("Auucracy of Test Data: {}%".format(100 * correct / total))
```
