```python
def draw_step(path, num_list):
    import csv
    import matplotlib.pyplot as plt
    
    path_list = []
    for item in num_list:
        path_list.append(path + " (" + str(item) + ").csv")
    
    
    f1 = open(path1, 'r', encoding = 'cp949')
    data1 = csv.reader(f1, delimiter = ',')
    header1 = next(data1)
    if num2 != 0:
        f2 = open(path2, 'r', encoding = 'cp949')
        data2 = csv.reader(f2, delimiter = ',')
        header2 = next(data2)

    step23 = []
    value23 = []
    step26 = []
    value26 = []

    for row in data1:
        step23.append( int( int(row[0].replace(',','')) / 1000) )
        value23.append( float(row[2].replace(',','')) )

    if num2 != 0:
        for row in data2:
            step26.append( int( int(row[0].replace(',','')) / 1000) )
            value26.append( float(row[2].replace(',','')) )

    y_label = str()
    if 'loss' in path:
        title = 'step_loss'
        y_label = 'loss'
    else:
        title = 'step_cer'
        y_label = 'cer'
    
    plt.suptitle(title,fontsize=20, x = 0.23)
    plt.tight_layout()
    plt.subplots_adjust(top=0.88)
    plt.xlabel('step (unit : 1k)', fontsize = 12)
    plt.ylabel(y_label, fontsize = 12)
    plt.plot(step23, value23, label = 'Kai #' + str(num1), color = 'skyblue')
    if num2 != 0:
        plt.plot(step26, value26, label = 'Kai #' + str(num2), color = 'hotpink')
        f2.close()
    plt.legend()
    plt.grid(True)
    plt.savefig(path + '.png')
    plt.show()
            
    f1.close()
```

```python
def draw_epoch(path, num1, num2 = 0):
    import csv
    import matplotlib.pyplot as plt
    
    path1 = path + " (" + str(num1) + ").csv" 
    path2 = path + " (" + str(num2) + ").csv" 
    
    f1 = open(path1, 'r', encoding = 'cp949')
    data1 = csv.reader(f1, delimiter = ',')
    header1 = next(data1)
    if num2 != 0:
        f2 = open(path2, 'r', encoding = 'cp949')
        data2 = csv.reader(f2, delimiter = ',')
        header2 = next(data2)

    step23 = []
    value23 = []
    step26 = []
    value26 = []

    for row in data1:
        step23.append( int(row[0].replace(',','')) )
        value23.append( float(row[2].replace(',','')) )

    if num2 != 0:
        for row in data2:
            step26.append( int(row[0].replace(',','')) )
            value26.append( float(row[2].replace(',','')) )
    
    y_label = str()
    if 'eval' in path :
        if 'loss' in path:
            title = 'eval_loss'
            y_label = 'loss'
        else:
            title = 'eval_cer'
            y_label = 'cer'
    else:
        if 'loss' in path:
            title = 'epoch_loss'
            y_label = 'loss'
        else:
            title = 'epoch_cer'
            y_label = 'cer'
        
    plt.suptitle(title,fontsize=20, x = 0.25)
    plt.tight_layout()
    plt.subplots_adjust(top=0.88)
    plt.xlabel('step', fontsize = 12)
    plt.ylabel(y_label, fontsize = 12)
    plt.plot(step23, value23, label = 'Kai #' + str(num1), color = 'skyblue')
    if num2 != 0:
        plt.plot(step26, value26, label = 'Kai #' + str(num2), color = 'hotpink')
        f2.close()
    plt.legend()
    plt.grid(True)
    plt.savefig(path + '.png')
    plt.show()
            
    f1.close()
```

```python
draw_step('scalar_train_step_loss', 22)
```

```python
draw_step('scalar_train_step_cer', 22)
```

```python
draw_epoch('scalar_train_epoch_loss', 22)
```

```python
draw_epoch('scalar_train_epoch_cer', 22)
```

```python
draw_epoch('scalar_eval_loss', 22)
```

```python
draw_epoch('scalar_eval_cer', 22)
```
