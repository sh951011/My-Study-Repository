```python
import librosa
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
```

```python
import sys
import wave
arg = 'KsponSpeech_000001'

with open(arg + '.pcm', 'rb') as pcmfile:
    pcmdata = pcmfile.read()
with wave.open(arg + '.wav', 'wb') as wavfile:
    wavfile.setparams((2, 2, 16000, 0, 'NONE', 'NONE'))
    wavfile.writeframes(pcmdata)
```

```python
%matplotlib inline

audio_path = 'KsponSpeech_000001.wav'

y, sr = librosa.load(audio_path, sr=16000)
time = np.linspace(0, len(y)/sr, len(y)) # time axis

fig, ax1 = plt.subplots() # plot
ax1.plot(time, y, color = 'pink', label='speech waveform')
ax1.set_ylabel("Amplitude") # y 축
ax1.set_xlabel("Time [s]") # x 축
plt.title(arg + 'wav') # 제목
plt.savefig(arg+'.png')
plt.show()
```
