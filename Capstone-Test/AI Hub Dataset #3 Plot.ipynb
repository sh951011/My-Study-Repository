```python
import pandas as pd
import matplotlib.pyplot as plt

loss1 = pd.read_csv("./csv/logmel_128 & encoder_5 & batct 8.csv", delimiter=',', encoding='cp949')["loss"]
cer1 = pd.read_csv("./csv/logmel_128 & encoder_5 & batct 8.csv", delimiter=',', encoding='cp949')["cer"]
loss2 = pd.read_csv("./csv/logmel_128 & encoder_5.csv", delimiter=',', encoding='cp949')["loss"]
cer2 = pd.read_csv("./csv/logmel_128 & encoder_5.csv", delimiter=',', encoding='cp949')["cer"]
loss3 = pd.read_csv("./csv/logmel_80 & encoder_5.csv", delimiter=',', encoding='cp949')["loss"]
cer3 = pd.read_csv("./csv/logmel_80 & encoder_5.csv", delimiter=',', encoding='cp949')["cer"]
```

```python
RGB = (0.4157, 0.2784, 0.3333)
plt.figure(figsize=(8,6))
plt.plot(loss1[1:15], color=RGB, label='Present (n_mels : 128, encoder : 5 batch : 8)')
plt.plot(loss2[1:15], label='Before (n_mels : 128, encoder : 5, batch : 5)')
plt.plot(loss3[1:15], label='Before (n_mels : 80, encoder : 5), batch : 6')
plt.legend(fontsize='x-large')
plt.title("A.I Hub Dataset #3 Loss", fontsize='x-large')
plt.xlabel("step (unit : 1000)", fontsize='x-large')
plt.ylabel("loss", fontsize='x-large')
plt.savefig("A.I Hub Dataset #2 Loss.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
RGB = (0.4157, 0.2784, 0.3333)
plt.plot(cer1[10:14], color=RGB, label='Present (n_mels : 128, encoder : 5 batch : 8)')
plt.plot(cer2[10:14], label='Before (n_mels : 128, encoder : 5, batch : 5)')
plt.plot(cer3[10:14], label='Before (n_mels : 80, encoder : 5), batch : 6')
plt.legend(fontsize='x-large')
plt.title("A.I Hub Dataset #3 cer", fontsize='x-large')
plt.xlabel("step (unit : 1000)", fontsize='x-large')
plt.ylabel("cer", fontsize='x-large')
plt.savefig("A.I Hub Dataset #3 cer.png")
plt.show()
```
