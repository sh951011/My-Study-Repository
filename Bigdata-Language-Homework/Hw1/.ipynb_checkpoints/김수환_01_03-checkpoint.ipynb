```python
def str_to_sentence(s):
    start=0      # 저장할 str의 시작 index
    end=0        # 저장할 str의 끝 index
    newS=""      # str 타입으로 ""로 초기화 및 선언
    data_list=[] # return 할 list 타입 초기화 및 선언
    
    for i in range(0,len(s),1) :   # s크기만큼 반복
        if(s[i] == '.' or s[i] == '!' or s[i] == '?'): # 만약 '.' '?' '!' 라면
            start+=1                 # start를 하나 증가하고,
            data_list.append(newS)   # 여태까지 newS에 추가해놓은 문자열을 data_list에 저장한다
            newS=""                  # 저장을 했으므로 newS는 ""로 초기화한다
            continue
        elif i==len(s)-1 :           # 마지막 반복이라면,
            data_list.append(newS)   # newS에 저장한 것을 추가하고 
            break                    # break
        else:                                  # '.' '?' '!' 가 아니라면
            end=i                              # 저장할 str의 끝을 end에 저장하고
            for j in range(start,end+1,1):     # s의 index 중 start~end를 반복하여
                newS+=s[j]                     # newS에 추가한다
            start=end+1                        # start를 하나 증가한다
            
    return data_list  
```

* 기능 : 문자열을 '.','?','!' (마침표들) 기준으로 문장별로 끊어서 list형태로 저장   

str을 char단위로 끊어서 하나하나 '.','?','!' 인지를 확인을 하여 발견되면 그 전의 마침표가 있었던   
문자 이후부터(start 변수) 방금 만난 마침표 직전(end 변수)까지를 newS(str type)에 저장하고 이 newS를   
data_list에 저장한다. 이런 식으로 저장을 하게 되면 data_list에 문장 단위로 저장이 된다. (마침표를 기준으로 분리)   

==> [[문장1],[문장2], ... , [문장n]]

```python
def change_to_space(data_list):
    space_list=[]    # return할 space_list 초기화 및 선언
    
    for i in range(0,len(data_list),1): # 문장 개수만큼 반복
        s=data_list[i]                  # 현재 진행하는 문장 일단 저장 -> for문이 반복될때마다 초기화 해야되므로 for문 안에 선언
        for j in range(0,len(s),1):          # 현재 문장의 크기만큼 반복
            if ord(s[j]) < 65 or (ord(s[j]) > 90 and ord(s[j]) < 97) or ord(s[j]) > 122 : # 알파벳 범위의 아스키 코드가 아니라면
                s=s.replace(s[j],' ')  # 해당 문자를 공백으로 변경한다
        space_list.append(s)           # 알파벳이 아닌 문자를 공백으로 변경한 문장을 space_list에 추가한다
    return space_list
```

* 기능 : 알파벳을 제외한 문자들을 공백으로 변경

위의 함수에서 텍스트를 마침표 기준으로 문장으로 분리하였으므로, 이 함수에서는 문장 단위로 검사를 하여,   
알파벳을 제외한 문자를 공백(' ')으로 변경한다.  

이를 위해서 문장의 개수만큼 반복 한다 ==> for i in range(0,len(data_list),1):    
이는 위에서 [[문장1],[문장2], ... , [문장n]] 와 같이 나누어졌다면 n번 반복하는 것을 의미한다 -> 문장별로 끊어서 확인한다.    

문장별로 끊어봐야 하므로, for문이 시작될 때마다 s라는 변수에 data_list[i] (i+1번째 문장) 를 저장을 먼저 한다.   
저장한 문장을 char 단위로 끊어본다. python에서는 문자의 ASCII CODE를 확인하려면 ord("A")와 같이 사용을 해야 하므로    
ord()함수를 이용해서 ASCII CODE로 변경을 하고, 해당 문자가 알파벳이 속해있는 ASCII CODE의 범위에 해당이 안되는지를 검사한다.    
알파벳이 속해있는 코드 범위가 아니라면, 알파벳이 아니란 의미이므로, 해당 문자를 공백으로 변경한다. (replace() 함수 사용)    
알파벳이 아닌 문자를 발견을 했던, 안했던, 내가 원하는 문장이 되었을 것이므로, 해당 문장을 space_llist에 추가한다.

```python
def split_by_space(data_list):     
    split_list=[]                            # 공백 기준으로 분리할 list 초기화 및 선언
    for i in range(0,len(data_list),1):  # 문장 개수만큼 반복
        s=data_list[i]         # 현재 문장 저장
        s=s.split(' ')         # 현재 문장 공백 기준으로 split (단어별로 저장)
        split_list.append(s)   # 분리한 단어 split_list에 추가
    
    return split_list
```

* 문자열을 공백 기준으로 분리해서 list 형식으로 저장

여태까지 한 과정을 살펴보면 먼저,    
[[문장1], [문장2] ... ,[문장n]] 으로 분리를 하고, 해당 문장들에서 알파벳이 아닌 문자는 공백(' ')으로 변경했다.    

이제 각 문장에는 공백(' ')이 가득할 것이다. 이 공백을 기준으로 문장을 단어로 분리하여 list(split_list)에 저장한다.     

Ex) The Jupyter Note*Book. 이라는 문장이 input으로 들어왔다면, 위의 3 함수를 거치게 되면,    
[[The], [Jupyter], [Note], [Book]] 으로 저장이 될 것이다.    

```python
def check_copy(original,compare):
    detect_list=[] # 최종적으로 return 할 list
    jump_i=0        # 중복적으로 검사하는 것을 방지하기 위해
    jump_j=0
    
    for ii in range(0,len(original),1): # original 문장 개수만큼 반복
        jump_i=0
        for jj in range(0,len(compare),1) :       # compare 문장 개수만큼 반복
            jump_j=0
            for i in range(0,len(original[ii]),1): # 문장의 단어 개수만큼 반복
                if i < jump_i :  # 중복적으로 검사하는 것을 방지하기 위함
                    continue    # 넘어감
                
                copy_list=[]         # copy_list 초기화
                for j in range(0,len(compare[jj]),1) : # 문장의 단어 개수만큼 반복
                    if j < jump_j :
                        continue
                    if original[ii][i] == compare[jj][j] :     # 같은 단어가 나온다면,
                        copy_list.append(original[ii][i])    # 단어를 copy_list에 추가하고
                        k=1
                        while True :  # 같은 단어가 안 나올 때까지 반복
                            if i+k >= len(original[ii]) or j+k >= len(compare[jj]) : # 문장 범위를 넘지않게 마지막 index에서는 멈추게 한다
                                break
                            if original[ii][i+k] == compare[jj][j+k] :        # 또 같은 단어가 나오면
                                copy_list.append(original[ii][i+k])       # copy_list에 추가한다
                                k+=1 # 하나하나 다음 단어와 비교하기 위한 변수
                            else :
                                break  # 다른 단어가 나오게 되면 break
                            
                if len(copy_list) > 3 :            # 크기가 4 이상이라면
                    detect_list.append(copy_list)   # 최종 list에 추가한다
                    jump_i=i+len(copy_list) # 지금 검사한 index i에 copy_list 크기를 더하면
                    jump_j=j+len(copy_list)
                                            # 다음에 검사할 list의 index가 된다
    return detect_list
```

* 기능 : 단어가 4개이상 겹치는지를 판별하여 4개이상 겹치는 문장들을 저장하여 list 형식으로 return 

실제 목표인 단어가 4개 이상 겹치는지를 판별하는 함수다.    
list 형식의 input인 original과 compare가 들어오게 되면, 각 문장들을 모두 대조해가며 비교를 해야 한다.    

Ex)
original : [[문장1], [문장2], ..., [문장n]]
compare : [[문장1], [문장2], ..., [문장m]]

위와 같은 2개의 텍스트를 비교를 한다면 이 n개의 문장을 m개의 문장과 하나하나 모두 대조를 해야 한다.
즉, n개를 m개만큼 비교한다는 의미이므로, 2중 for문으로 구현 가능하다.   

    
    for ii in range(0,len(original),1): # original 문장 개수만큼 반복
        for jj in range(0,len(compare),1) :       # compare 문장 개수만큼 반복
        
위의 구조에 해당하는 부분이다. 각 문장들을 대조하는 구조는 갖춰졌으므로, 
이 구조 안에서 각 문장안의 단어들을 모두 비교해야 한다.    

각 문장들은 아래와 같은 구조를 가진다.    

[[단어1], [단어2], ..., [단어n]]   
[[단어1], [단어2], ..., [단어m]]   

이도 위와 똑같이 구현을 해주면 된다. n개의 단어를 m개의 단어와 하나하나 모두 대조를 해야하므로,     
이도 2중 for문으로 구현가능하다.    

            for i in range(0,len(original[ii]),1): # 문장의 단어 개수만큼 반복
                for j in range(0,len(compare[jj]),1) : # 문장의 단어 개수만큼 반복

위의 구조가 단어를 비교하는 구조이다.

하여, 총 구조를 보면 4중 for문을 가지는 구조를 가지게 된다.     

이 구조 안에서 각 단어를 비교를 해가다가, 같은 단어가 나오게 된다면 -> if original[ii][i] == compare[jj][j] :    

while문으로 진입을 하게 된다.    
왜 while문이나면, 4단어가 아닌, 2단어, 3단어, 아니면 더 많은 단어까지도 나올 수 있으므로, 다른 단어가 나올 때까지 반복해야하므로,    
while문을 사용.    

그리고, 한 단어가 같던지, 10단어가 같던지 모두 copy_list에 저장을 한다. (copy_list는 새로운 문장과 비교를 할 때 초기화를 한다)    

이렇게 copy_list에 저장을 한 단어들은 if문을 마주치게 된다. -> if len(copy_list) > 3 :   

이 if문은 copy_list에 4개 이상의 단어가 저장이 되어 있을 때만 진입이 가능하다.   
즉, 4개의 단어 이상이 저장이 되어있다면, detect_list(최종적으로 return할 리스트)에 추가(append)를 한다.   
이렇게 list형식으로 저장한 copy_list를 detect_list에 추가를 한다면, list에 list에 추가한 것이므로, 한번에 겹쳤던 단어들 모음으로   
저장을 할 수 있다는 장점이 있다.

* jump_i & jump_j의 역할    

The jupyter Notebook is ~ that 문장과    
The jupyter Notebook is ~ which 문장을 비교한다고 생각해보자.    

내가 짠 위의 구조에서 아무것도 안하고 코드를 돌린다면, detect_list에는

와 같이 앞 단어를 하나씩 뺀채로 저장이 될 것이다. (jump_i와 jump_j 빼고 돌려보면 확인가능.)    

이와 같은 일을 방지하기 위해, detect_list에 저장을 하게 될 경우(4개 이상 단어가 겹쳤을 경우)      

jump_i=i+len(copy_list)   
jump_j=j+len(copy_list)    

를 해준다. 즉, 일정 개수 이상 단어가 겹쳐서 저장을 했다면, 이미 검사가 끝난 단어들이므로, jump를 해준다.    
-> 각 for문 시작할 때 if문으로 jump 변수로 확인해서 continue 해주는 이유      

```python
def list_to_str(data_list) :
    s=""
    for i in range(0,len(data_list),1):
        s+="\n"+str(i+1)+" "
        for j in range(0,len(data_list[i]),1):
            s+=(data_list[i][j]+" ")

    return s
```

* 기능 : list 형식으로 저장한 겹치는 문장들을 사람이 보기 편하도록 문자열 형식으로 저장하여 return 하는 함수

최종적으로 표절을 검사한 문장들은 list형식으로 사람들이 보기에는 불편하게 되어 있다.   
이를 사람이 보기 편한 str 형식으로 살짝 꾸며서 return 해준다.

```python
text1=open('c:/app/d1.txt','r') # 텍스트 저장
text2=open('c:/app/d2.txt','r') # 텍스트 저장

str1=text1.read() # text to str
str2=text2.read() # text to str

text1.close() # fclose
text2.close() # fclose

print("\n【비교문장\n")

data_list1=str_to_sentence(str1) # 문자열을 '.' '?' '!' 기준으로 끊어서 저장
data_list2=str_to_sentence(str2) # 문자열을 '.' '?' '!' 기준으로 끊어서 저장

print(data_list1)

print("## Original\n\n"+str1)
print("\n## Compare\n\n"+str2)

data_list1=change_to_space(data_list1) # 알파벳을 제외한 문자들을 공백으로 변경
data_list2=change_to_space(data_list2) # 알파벳을 제외한 문자들을 공백으로 변경

data_list1=split_by_space(data_list1) # 공백을 기준으로 단어 단위로 저장
data_list2=split_by_space(data_list2) # 공백을 기준으로 단어 단위로 저장

copy_list=check_copy(data_list1,data_list2) # 2개의 문장을 비교하여 4단어 이상 연속된 단어들을
                                            # 연속된 단어별로 끊어서 저장
                                            # ~~1~~ that ~~2~~ 이라면 [[1],[2]] 와 같은 형식으로 저장

s=list_to_str(copy_list) # list 형식으로 되어있는 copy_list를 보기 편한 문자열로 변경 
print("\n【실행결과")
print(s) 
```
