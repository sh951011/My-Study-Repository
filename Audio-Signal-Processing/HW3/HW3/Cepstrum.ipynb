```python
import math
import matplotlib.pyplot as plt
# FILE READ
forigin = open("spectrum.txt","rt")
flenv = open("low_freq_envelope.txt","rt")
fhenv = open("high_freq_envelope.txt","rt")
flowlifter = open("low_filterred.txt","rt")

# SCALE
lenv_scale = 1/240
spec_scale = 1

# GET DATA
lenv = [float(spec_env_val.replace('\n','')) for spec_env_val in flenv.readlines()]
henv = [float(spec_env_val.replace('\n','')) for spec_env_val in fhenv.readlines()]
spec = [float(spec_env_val.replace('\n','')) for spec_env_val in forigin.readlines()]
log_spec = [math.log10(float(spec_env_val.replace('\n',''))) for spec_env_val in forigin.readlines()]
lowlifter = [float(spec_env_val.replace('\n','')) for spec_env_val in flowlifter.readlines()]

scaled_lenv = [spec_element * 0.004 for spec_element in lenv]
scaled_henv = [spec_element * lenv_scale for spec_element in henv]
scaled_spec = [spec_element * spec_scale for spec_element in spec]
log_scaled_spec = [math.log10(spec_element) * spec_scale for spec_element in spec]

# PLOT
# plt.plot(lenv[:160])
# plt.plot(scaled_henv[:160])
# plt.plot(log_spec[:160])
plt.figure(figsize=(10, 6))
plt.title('Cepstral - log spectrum', fontsize='xx-large')
plt.xlabel('n', fontsize='x-large')
plt.ylabel('log spectrum', fontsize='x-large')
plt.plot(henv[:160], color='skyblue')
plt.show()
# plt.plot(spec[:])
# plt.plot(scaled_spec[:])
# plt.plot(lowlifter[0:160])
# plt.plot(log_scaled_spec[:])

# 예비 정답
plt.figure(figsize=(10, 6))
plt.title('Cepstral - low time envelope', fontsize='xx-large')
plt.xlabel('n', fontsize='x-large')
plt.ylabel('log spectrum', fontsize='x-large')
plt.plot(log_scaled_spec[:160], color='skyblue')
plt.plot(scaled_lenv[:160], color='red')

# FILE CLOSE
flenv.close()
fhenv.close()
forigin.close()
flowlifter.close()
```
